# Diffxpy & PAGA

In this notebook, we aim to accomplish two objectives:

1. Conduct a differential expression analysis with diffxpy to find the markers of the different subclones in chronic lymphocitic leukemia (CLL) patients introducing cold-shock score as a covariate.
2. Infer the trajectories of the evolution of the aforementioned clonality across the disease course using PAGA.

## Pre-processing

### Import packages

### Load data

We will load an [anndata](https://icb-anndata.readthedocs-hosted.com/en/stable/anndata.AnnData.html) object that was converted from our previously processed [Seurat](https://satijalab.org/seurat/) object using [sceasy](https://github.com/cellgeni/sceasy). The object contains the cells from the two CLL subclones in the patient ICGC_012, which showed a gradual expression of CXCR4, a gene that is essential in the lymph node microenvironment. We hypothesized that after diagnosis, two different clones evolve depending on their exposure to the lymph node.

```python
adata = sc.read_h5ad("results/anndata_objects/richter_012_2_subclones.h5ad")
adata
```

```python
sc.pl.umap(adata, color = "seurat_clusters", size = 100)
```

## Diffxpy

https://www.youtube.com/watch?v=Ib8U01APqeE



```python
import diffxpy.api as de
```

```python
det = de.test.wald(
    adata = adata,
    formula = "~1+cluster+nCount_RNA+cold_shock_score11",
    fac_to_test = "cluster",
    noise_model = "negative_binomial",
    training_strategy = "IRLS"
)
```

```python
det.qval()
```

```python
det.summary()
```

```python
det.volcano()
```
